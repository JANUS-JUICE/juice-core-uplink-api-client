# All together in a plot

```python
plt.figure(figsize=(18,5))

import numpy as np


ring_sc_elevation.plot(ax=plt.gca(), linestyle='dotted')


for (label, group), c in zip(eu_flybys.groupby('name'), ['k','r', 'g', 'b']):
    for  j, row in group.iterrows():
        p1 = plt.axvline(row.start, label=label, color=c   )
        
for i, seg in plan.query(" segment_definition == 'JA_PE12'" ).iterrows():
    plt.axvspan(seg.start, seg.end, alpha=0.1, color= 'k', label= 'JA_PE12 segments')

handles, labels = plt.gca().get_legend_handles_labels()
labels, ids = np.unique(labels, return_index=True)


plt.legend(np.array(handles)[ids], labels, fontsize= 'x-small')


plt.xlim(plan.start.min() - np.timedelta64(60, 'D'), plan.end.max() + np.timedelta64(60, 'D'))
```
